```python
import pandas as pd
from pathlib import Path
```

```python
# File to Load 
food_event_df = Path("food_df.csv")
```

```python
#Read Food Data File and store into Pandas DataFrames
food_df = pd.read_csv(food_event_df)
food_df.head()
```

```python
# Convert the 'date_started' column to datetime
food_df['date_started'] = pd.to_datetime(food_df['date_started'], format='%Y%m%d')

# Extract the year and create a new column
food_df['Year'] = food_df['date_started'].dt.year

# Assuming 'df' is your DataFrame and 'year' is the column with the year
food_df['Year'] = food_df['Year'].astype(str).str.split('.').str[0].astype(int, errors="ignore")
food_df.head()


```

```python
#Value count for Genders affected
total_affected = food_df["consumer_gender"].value_counts()
print(total_affected)
```

```python
#Value count for industry Name
foods_affected = food_df["industry_name"].value_counts()
print(foods_affected)
```

```python
#Value count for industry code
indus_code_affected = food_df["industry_code"].value_counts()
print(indus_code_affected)
```

```python
role_tag_affected = food_df["role"].value_counts()
print(role_tag_affected)
```

```python
#Create new dataframe for products that resulted in Death
death_count = food_df['reactions'].str.contains('death').sum()
print(death_count)
death_df = food_df[food_df['reactions'].str.contains('death')]
death_df.head()
```

```python
#total count of all Null values across entire dataset ( every column )
null_count = food_df.isnull().sum()
print(f"The total amount of null and NaN values in the column is {null_count}")
```

```python
#Create new dataframe for products that resulted in Cancer
cancer_count = food_df['reactions'].str.contains('cancer').sum()
print(cancer_count)
cancer_df = food_df[food_df['reactions'].str.contains('cancer')]
cancer_df.head()
```
