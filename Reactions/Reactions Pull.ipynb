```python
import pandas as pd
```

```python
# Save and read data files
reactions_data = "reactions_detailed.csv"

reactions_detailed = pd.read_csv(reactions_data)

reactions_detailed
```

```python
# Find the top 5 industry names.
reactions_detailed['industry_name'].value_counts()
```

```python
# Drop Cosmetics from industry_name.
reactions_detailed = reactions_detailed.loc[reactions_detailed['industry_name']!='Cosmetics']

reactions_detailed
```

```python
# Filter out the top 5 industry names.

industry_top_5 = reactions_detailed.loc[
    (reactions_detailed['industry_name'] == 'Vit/Min/Prot/Unconv Diet(Human/Animal)')|
     (reactions_detailed['industry_name'] == 'Nuts/Edible Seed')|
      (reactions_detailed['industry_name'] == 'Vegetables/Vegetable Products')|
       (reactions_detailed['industry_name'] == 'Soft Drink/Water')|
        (reactions_detailed['industry_name'] == 'Bakery Prod/Dough/Mix/Icing')]

industry_top_5
```

```python
# Clean up the dataset and drop select columns.

industry_all_pro_team = industry_top_5.drop(
    columns=[
        'Unnamed: 0',
        'report_number',
        'consumer_gender',
        'date_created',
        'death?',
        'hospitalization?',
        'reactions',
        'diarrhea?'
        ])

industry_all_pro_team
```

```python
# Locate rows that are associated to diarrhea.

diarrhea = reactions_detailed.loc[reactions_detailed['diarrhea?']==True]

diarrhea
```

```python
# Locate the top 5 industry names that are associated to diarrhea.

diarrhea['industry_name'].value_counts()
```

```python
# Filter out the top 5 industry names associated to diarrhea.

diarrhea_top_5 = diarrhea.loc[
    (diarrhea['industry_name'] == 'Vit/Min/Prot/Unconv Diet(Human/Animal)')|
     (diarrhea['industry_name'] == 'Nuts/Edible Seed')|
      (diarrhea['industry_name'] == 'Vegetables/Vegetable Products')|
       (diarrhea['industry_name'] == 'Fishery/Seafood Prod')|
        (diarrhea['industry_name'] == 'Fruit/Fruit Prod')]

diarrhea_top_5
```

```python
# Clean up the dataset and drop select columns.

diarrhea_all_pro_team = diarrhea_top_5.drop(
    columns=[
        'Unnamed: 0',
        'report_number',
        'consumer_gender',
        'date_created',
        'death?',
        'hospitalization?',
        'reactions'
        ])

diarrhea_all_pro_team
```

```python
diarrhea_all_pro_team.groupby("Year")["industry_name"].value_counts()
```
