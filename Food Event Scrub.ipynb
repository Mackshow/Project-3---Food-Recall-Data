```python
# Save and read data files
food_event_data = "food_df.csv"

food_df = pd.read_csv(food_event_data)

food_df.head()
```

```python
#Check data info.
food_df.info()
```

```python
#Rename consumer.age, consumer.age_unit, consumer.gender columns.
food_df = food_df.rename(
    columns={
        'consumer.age':'consumer_age',
        'consumer.age_unit':'consumer_age_unit',
        'consumer.gender':'consumer_gender'})

food_df.head(1)
```

```python

# Convert the 'date_started' column to datetime
food_df['date_created'] = pd.to_datetime(food_df['date_created'], format='%Y%m%d')
# Extract the year and create a new column
food_df['Year'] = food_df['date_created'].dt.year
# Assuming 'df' is your DataFrame and 'year' is the column with the year
food_df['Year'] = food_df['Year'].astype(str).str.split('.').str[0].astype(int, errors="ignore")
food_df.head()
```

```python
# Drop columns not being used for overall project.
report_number_dates_created = food_df.drop(
    columns=[
        'role',
        'industry_code',
        'consumer_age',
        'consumer_age_unit',
        'date_started'
        ])

report_number_dates_created
```

```python
#Split up the reactions column.
reactions_split_1 = food_df['reactions'].str.split(
    pat="[",
    expand=True)

reactions_split_2 = reactions_split_1[1].str.split(
    pat="]",
    expand=True)

reactions_split = reactions_split_2[0].str.split(
    pat=",",
    expand=True)

reactions_split
```

```python
# Find which rows have the food reaction diarrhea and apply the answer to a column new column.
reactions_split['diarrhea?'] = reactions_split.iloc[:, 0:70].apply(lambda x: x.str.contains('DIARRHEA')).any(axis=1)
reactions_split['death?'] = reactions_split.iloc[:, 0:70].apply(lambda x: x.str.contains('DEATH')).any(axis=1)
reactions_split['hospitalization?'] = reactions_split.iloc[:, 0:70].apply(lambda x: x.str.contains('Hospitalization')).any(axis=1)
```

```python
# Merge reaction data with revised report.
reactions_detailed = pd.merge(
    report_number_dates_created,
    reactions_split,
    left_index=True,
    right_index=True)

# Drop outcomes column
reactions_detailed = reactions_detailed.drop(columns='outcomes')
reactions_detailed
```

```python
# Test a search for diarrhea in each column.
test = reactions_detailed.loc[reactions_detailed['diarrhea?'] == True]

test
```

```python
# Transport reactions to CSV file.
reactions_detailed.to_csv("Reactions/reactions_detailed.csv")
```
